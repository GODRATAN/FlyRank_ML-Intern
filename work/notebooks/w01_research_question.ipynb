## 3. Quick look at the data (2-3 real numbers)

```python
import os
import sys
import subprocess
import pandas as pd

REPO_URL = "https://github.com/flyrank-bih/flyrank-ml-internship-starter"
REPO_DIR = "flyrank-ml-internship-starter"

if not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

os.chdir(REPO_DIR)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(df.shape)
```

```python
# Supporting numbers from the starter dataset

visible = df[df["impressions_90d"] > 0]
corr = visible["search_volume"].corr(visible["impressions_90d"])

print(f"Correlation (search volume vs impressions): {corr:.3f}")

print(f"Total pages in dataset: {len(df)}")

declining = (df["trend_direction"].str.lower() == "down").sum()
print(f"Pages marked as declining: {declining}")
```

### Observation

The starter dataset contains **30,000** content pages, of which **16,262** are marked as declining. After filtering to pages with impressions greater than zero, the correlation between search volume and impressions is **0.001**, indicating almost no linear relationship. These observations suggest that relying on search volume alone may not be sufficient for prioritizing content refreshes, making it worthwhile to investigate multiple signals during the internship.
